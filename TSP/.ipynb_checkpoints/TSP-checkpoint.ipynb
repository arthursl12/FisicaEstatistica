## Algoritmos

### Imports e Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
```

A fim de melhor ilustrar o desenvolvimento do algoritmo, é interessante ter uma função que plota tanto a localização espacial das cidades no plano quanto o caminho proposto.

```python
def plotPath(path, x, y, dist_list=None, title=None):
    """
    Plota as cidades e o caminho passado.
    Se a evolução da distância total for dada também, será plotada ao lado.
    
    Parâmetros:
        path: caminho proposto
        x: coordenadas x das cidades
        y: coordenadas y das cidades
        dist_list: (opcional) evolução da distância entre os passos de Monte Carlo
        title: (opcional) título para ser colocado sobre o(s) plot(s)
    Retorno:
        None
    """
    if (dist_list is None):
        fig, ax1 = plt.subplots(figsize=(6,6))
        ax2 = None
    else:
        fig, (ax1, ax2) = plt.subplots(1,2, figsize=(13,6))
    
    
    # Plot cidades
    # Início em verde, fim em vermelho
    # Intermediárias em azul
    ax1.scatter(x[path[0]], y[path[0]], color='green',
                edgecolors='black',linewidths=.5, s=150)
    ax1.scatter(x[path[1:-1]], y[path[1:-1]], color='blue',
                edgecolors='black', linewidths=.5)
    ax1.scatter(x[path[-1]], y[path[-1]], color='red',
                edgecolors='black', linewidths=.5, s=150)
    
    # Plot do caminho
    path_x = []
    path_y = []
    for city_idx in path:
        path_x.append(x[city_idx])
        path_y.append(y[city_idx])
        
    # Voltar ao início
    path_x.append(x[path[0]])
    path_y.append(y[path[0]])
    ax1.plot(path_x, path_y)
    
    if (ax2 is not None):
        # Plot evolução distância
        ax2.plot(dist_list)
        ax2.grid()
    fig.suptitle(title, y=0.92)
```

### Funções Auxiliares

Para fazer as simulações, precisamos de algumas funções para cálculos auxiliares, a fim de uma melhor modularização do código. Primeiramente, é muito útil pré-computar as distâncias entre todas as cidades a fim de evitar cálculos desnecessários.

```python
def distTable(x_ct,y_ct):
    """
    Pré-calcula as distância entre todas as cidades e as guarda em uma matriz.
    Aqui a dist(i->j) = dist(j->i), i.e., a matriz é simétrica com relação à 
    diagonal principal
    
    Parâmetros:
        x_ct: coordenadas x das cidades
        y_ct: coordenadas y das cidades
    Retornos:
        tab: matriz de distâncias
    """
    N = len(x_ct)
    tab = np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            t1 = (x_ct[i]-x_ct[j])**2
            t2 = (y_ct[i]-y_ct[j])**2
            dist = np.sqrt(t1+t2)
            tab[i,j] = dist
            tab[j,i] = dist
    return tab
```

Além disso, temos que saber calcular quanto de energia um determinado caminho tem, ou seja, qual a distância total percorrida.

```python
def energyPath(path, dists):
    """
    Calcula a energia de um caminho, dada a tabela de distâncias.
    Algoritmo do enunciado.
    
    Parâmetros: 
        path: caminho, com índices das cidades na ordem percorrida
        dists: tabela de distâncias entre elas
    Retornos:
        ener: energia do caminho
    """
    N = len(path)
    ener = 0
    for i in range(N-1):
        ct1_idx = path[i]
        ct2_idx = path[i+1]
        ener += dists[ct1_idx][ct2_idx]
        
    # Após a última, voltamos ao início
    ct1_idx = path[N-1]
    ct2_idx = path[0]
    ener += dists[ct1_idx][ct2_idx]
    return ener
```

Durante o algoritmo, temos que propor alterações no caminho atual a fim de continuar nossa busca no espaço de configurações. Usando a heurística proposta no enunciado, de inverter um intervalo do caminho, vamos sugerir um caminho alternativo.

```python
def genNewPath(path):
    """
    Gera um novo caminho ao inverte a ordem entre dois pontos do caminho,
    aleatoriamente.
    Adaptado do algoritmo do enunciado.
    
    Parâmetros:
        path: caminho, com índices das cidades na ordem percorrida
    Retornos:
        npath: caminho com inversão feita
        i: primeiro índice do intervalo de inversão
        j: último índice do intervalo de inversão
    """
    N = len(path)
    npath = np.zeros(N)
    
    # Escolhe i e j aleatóriamente, com i != j
    i = np.random.randint(N)
    j = i
    while (j == i): j = np.random.randint(N)
    
    # Inverte i e j se i>j
    if (i > j):
        tmp = i
        i = j
        j = tmp
    
    # Inverte os elementos da lista entre i e j (inclusos)
    npath = path.copy()
    sublist = npath[i:j+1]
    sublist.reverse()
    npath[i:j+1] = sublist
    return npath, i, j
```

Finalmente, para conseguir avaliar esse novo caminho, sem ter que fazer várias contas, vamos usar o algoritmo proposto no enunciado a fim de calcular apenas a variação da energia (ou seja, da distância total percorrida) com a alteração do caminho antigo para o novo.

```python
def deltaEnergy(path, npath, dists, ini, fim):
    """
    Calcula o delta de energia com a mudança entre os caminhos propostos,
    a qual deve ser apenas a inversão da ordem entre os índices 'ini' e 
    'fim'. Algoritmo do enunciado.
    
    Parâmetros:
        path: caminho antigo (sem inversão)
        npath: caminho novo (com inversão)
        dists: matriz de distâncias
        ini: índice inicial da inversão
        fim: índice final da inversão (incluso)
    Retornos:
        de: delta de energia após mudança
    """
    # Descobrindo a cidade antes da inversão
    lft = ini - 1
    if (lft < 0): lft = N - 1
        
    # Descobrindo a cidade logo após a inversão
    rgt = fim + 1
    if (rgt > N - 1): rgt = 0
    
    # Calculando o delta com a fórmula fornecida
    de = 0
    # economizamos a distância antiga 
    de -= dists[path[lft]][path[ini]]
    de -= dists[path[fim]][path[rgt]]
    # custou a nova distância
    de += dists[npath[lft]][npath[ini]]
    de += dists[npath[fim]][npath[rgt]]
    
    return de
```

### Algoritmo de Metropolis

Embora já implementado em atividades anteriores, achou-se melhor adaptá-los em uma nova implementação que necessariamente usar os já feitos. Assim, temos aqui uma nova implementação do passo de Monte Carlo.

```python
def mcstep(path, temp, dists, ener):
    """
    Executa um passo de Monte Carlo: tenta inverter entre alguns índices,
    aceita se diminuir a energia ou se estiver dentro de uma probabilidade.
    
    Parâmetros:
        path: caminho atual
        temp: temperatura a ser considerada (não é o beta)
        dists: matriz de distâncias
        ener: energia atual
    Retornos:
        ener: energia pós passo de Monte Carlo (modificada ou não)
        path: caminho pós passo de Monte Carlo (modificado ou não)
    """
    
    npath, ini, fim = genNewPath(path)
    de = deltaEnergy(path, npath, dists, ini, fim)
    if (de < 0):
        # se diminuir a energia, aceite
        path = npath
        ener += de
    elif(np.random.random() < np.exp(-de/temp)):
        # se estiver na probabilidade, aceite
        path = npath
        ener += de
    
    return ener, path
```

Em seguida, temos a adaptação do algoritmo de Metropolis que faz, por padrão, como recomendado pelo enunciado, 100 passos de Monte Carlo em uma dada temperatura.

```python
def metropolis(path, dists, temp, steps=100):
    """
    Executa 'steps' passo de Monte Carlo para um TSP de tamanho N à temperatura 'temp'.
    O caminho inicial usado será o fornecido. 
    
    
    Parâmetros:
        path: caminho inicial
        dists: matriz de distâncias
        temp: temperatura a ser considerada (não é o beta)
        steps: (opcional) quantidade de passos de Monte Carlo a serem dados (padrão=100)
    Retornos:
        ener: distância (energia) ao final de todos passos
        path: caminho proposto no restante

    """
    ener = energyPath(path, dists)
    for i in range(steps):
        ener, path = mcstep(path, temp, dists, ener)
        
    return ener, path
```

### Funções para Simulações

Para simularmos o TSP e encontrarmos resultados, temos que partir de uma temperatura inicial e resfriar o sistema segundo uma política até uma temperatura final mais baixa. Desse modo, com o uso do Algoritmo de Metropolis, espera-se encontrar um caminho mínimo ao final do resfriamento.

```python
def simulate(N, x, y,
             Tmax, Tmin, cte,
             initial_path=None):
    """
    Encontra um caminho mínimo (local ou global) do TSP a partir do algoritmo de Metropolis,
    começamos da temperatura Tmax e diminui-se linearmente até a mínima.
    
    Parâmetros:
        N: quantidade de cidades
        x: coordenadas x das cidades
        y: coordenadas y das cidades
        Tmax: temperatura máxima, i.e., temperatura inicial
        Tmin: temperatura mínima, i.e., temperatura final
        cte: constante de resfriamento
        initial_path: caminho inicial. 
                      Por padrão, são as cidades em ordem crescente de índice
    Retornos:
        path: caminho de distância mínima encontrada
        ener: distância (energia) desse caminho
        dist_list: evolução da distância (energia) ao longo do resfriamento
    """
    # Inicializa estruturas auxiliares da rede
    dists = distTable(x, y)
    if (initial_path is None): 
        path = list(range(N))
    else: 
        path = initial_path
    ener = energyPath(path, dists)
    dist_list = []
    dist_list.append(ener)
    
    T = Tmax
    while(T > Tmin):
        ener, path = metropolis(path, dists, temp=T, steps=100)
        dist_list.append(ener)
        T = T * cte
        
    return path, ener, dist_list
```

Contudo, não temos garantia de que o caminho final encontrado é, de fato, o mínimo global ou apenas um mínimo local. Assim, é necessário fazer várias amostras partindo de configurações iniciais diferentes a fim de se chegar mais próximo do mínimo global. Para melhor visualizar a evolução ao longo das diferentes amostras, sempre que encontramos um caminho melhor, ele é plotado.

```python
def simulateSamples(N, x, y,
                    Tmax, Tmin, cte,
                    samples=10, plots=False):
    """
    Encontra um caminho mínimo (local ou global) do TSP a partir de várias amostras 
    de simulações usando o algoritmo de Metropolis. Começamos da temperatura Tmax e
    diminui-se linearmente até a mínima. 
    
    Parâmetros:
        N: quantidade de cidades
        x: coordenadas x das cidades
        y: coordenadas y das cidades
        Tmax: temperatura máxima, i.e., temperatura inicial
        Tmin: temperatura mínima, i.e., temperatura final
        cte: constante de resfriamento
        samples: quantidade de amostras. Por padrão são 10.
    Retornos:
        path: caminho de distância mínima encontrada
        ener: distância (energia) desse caminho
        dist_list: evolução da distância (energia) ao longo do resfriamento
    """
    # Um caminho aleatório como mínimo apenas para guardar lugar
    path_min = list(range(N))
    np.random.shuffle(path_min) 
    ener_min = energyPath(path_min, distTable(x,y))
    dist_list_min = None
    
    for i in range(samples):
        # Caminho aleatório inicial a cada amostra
        path = list(range(N))
        np.random.shuffle(path)
        path, ener, dist_list = simulate(N, x, y, Tmax, Tmin, 
                                         cte, initial_path=path)
        
        # Atualizar o que temos se o encontrado for melhor
        if (ener < ener_min):
            path_min = path
            ener_min = ener
            dist_list_min = dist_list
            if (plots):
                plotPath(path, x, y, dist_list, f"Amostra {i+1}, d={round(ener,2)}")
                plt.show()
    return path_min, ener_min, dist_list_min
```

## Resultados

### Caminho Circular

Para ilustrar de início, vamos tomar um caminho circular de cidades. Isso nos facilita pois sabemos que o caminho mais curto é com certeza dar a volta no círculo sempre seguindo para a cidade adjacente. Dessa forma, podemos aferir o quão próximo do ideal chegou o algoritmo dados diferentes parâmetros e políticas de seguimento.

Para dispor a cidade em círculos, vamos usar uma função dedicada, obtida de uma das referências auxiliares do enunciado.

```python
def circle(N):
    """
    Cria N cidades dispostas num círculo.
    Fonte: 
    https://nathanrooy.github.io/posts/2020-05-14/simulated-annealing-with-python/
    
    Parâmetros:
        N: quantidade de cidades
    Retorno:
        x: coordenadas x das cidades 
        y: coordenadas y das cidades
    """
    dr = (2 * np.pi) / N

    x = []
    y = []
    for i in range(N):
        radians = dr * i
        x.append(N*dr+np.cos(radians))
        y.append(N*dr+np.sin(radians))
    return np.array(x),np.array(y)
```

Para ilustrar, vamos dispor as cidades e o caminho seguindo as cidades em ordem crescente de índice. Dado o funcionamento do algoritmo, seguir em ordem de índice significa sempre seguir para cidades adjacentes no círculo, o que culmina, como já comentado, no caminho mais curto. 

```python
N = 60
x,y = circle(N)
path = list(range(N))
```

```python
plotPath(path, x, y)
plt.title(f"d={round(energyPath(path, distTable(x,y)),2)}")
plt.show()
```

Naturalmente, cada par de vértices adjacentes gera um caminho análogo. Mas o importante é que o mais curto sempre segue a borda do círculo.

#### Mudança no Resfriamento

Se mudarmos a taxa de resfriamento, o resultado é um pouco diferente.

```python
simulateSamples(N, x, y, Tmax=10, Tmin=0.005, cte=0.95, samples=10, plots=True);
```

### Caminho aleatório pequeno

Vamos observar o comportamento para uma organização de cidades aleatória, com o mínimo delas, ou seja, 10 cidades apenas.

```python
N = 10
x,y = np.random.random(N), np.random.random(N)
path = list(range(N))
```

```python
plotPath(path, x, y)
plt.title(f"d={round(energyPath(path, distTable(x,y)),2)}")
plt.show()
```

No caso mais "desfavorável", vemos que ainda assim, ele consegue chegar num caminho mínimo logo na primeira amostra.

```python
simulateSamples(N, x, y, Tmax=5, Tmin=0.005, cte=0.8, samples=10, plots=True);
```

```python
path, ener, dist_list = simulateSamples(N, x, y, Tmax=5, Tmin=0.005, cte=0.95, samples=50, plots=False)
plotPath(path, x, y, dist_list, title=f"d={round(ener,2)}")
plt.show()
```

Com mais amostras e com menor taxa de resfriamento, o resultado é o mesmo. Isso nos leva a observar que para sistemas com poucas cidades, muitas amostras ou então taxas muito altas apenas precisarão de mais tempo computacional, sem necessariamente obter resultados mais próximos do mínimo global.  

Os gráficos de evolução da distância durante o algoritmo de Metropolis também acusam algo semelhante. Num sistema menor, a diferença a cada passo de Monte Carlo fica bastante grande. Com mais iterações,vê-se que a todo momento o sistema vai para um caminho demasiado longo.

Pode ser fruto da coincidência, mas, muitas vezes, o caminho mais curto não possui arestas que se cruzam, como foi o caso.

### Caminho aleatório mediano

```python
path, ener, dist_list = simulateSamples(N, x, y, Tmax=5, Tmin=0.005, cte=0.8, samples=10, plots=False);
plotPath(path, x, y, dist_list, title=f"d={round(ener,2)}")
plt.show()
```

A solução parece razoável. Contudo, vamos comparar com uma solução com uma taxa menor e mais amostras.

Ainda há uma melhora, notam-se algumas diferenças nos caminhos obtidos. Há de se registrar aqui o tempo significativamente maior que essa taxa exigiu.

### Caminho aleatório grande

Com 60 cidades, não temos todo o luxo de poder checar muitas possibilidades com uma taxa de resfriamento tão lenta.

```python
N = 60
x,y = np.random.random(N), np.random.random(N)
```

```python
path, ener, dist_list = simulateSamples(N, x, y, Tmax=5, Tmin=0.005, cte=0.8, samples=10, plots=False);
plotPath(path, x, y, dist_list, title=f"d={round(ener,2)}")
plt.show()
```

```python
path, ener, dist_list = simulateSamples(N, x, y, Tmax=10, Tmin=0.005, cte=0.8, samples=50, plots=False);
plotPath(path, x, y, dist_list, title=f"d={round(ener,2)}")
plt.show()
```

Apenas aumentar a quantidade de amostras não é suficiente aqui. Há tantas possibilidades de caminho que algumas vezes, uma execução com mais amostras retorna um caminho mais longo que outras com menos amostras, o que não é comum.

```python
path, ener, dist_list = simulateSamples(N, x, y, Tmax=10, Tmin=0.0001, cte=0.8, samples=50, plots=False);
plotPath(path, x, y, dist_list, title=f"d={round(ener,2)}")
plt.show()
```

Nesse caso com várias cidades, os limites de temperatura são essenciais para bons resultados. Isso não é tão perceptível com problemas menores. Uma temperatura inicial alta nos permite explorar o espaço melhor e uma final menor nos permite chegar mais perto (talvez até mesmo alcançar) os mínimos. Convém notar que a taxa de resfriamento está no nível mínimo recomendado: 0.8, denotando uma queda mais rápida

```python
path, ener, dist_list = simulateSamples(N, x, y, Tmax=5, Tmin=0.005, cte=0.95, samples=10, plots=False);
plotPath(path, x, y, dist_list, title=f"d={round(ener,2)}")
plt.show()
```

Num outro cenário, em que temos temperaturas desfavoráveis e poucas amostras, mas uma taxa de resfriamento mais lenta, conseguimos melhores resultados. No caso, a um custo de mais iterações no algoritmo de Metropolis.

## Conclusões

Após bastantes simulações, é possível tirar algumas conclusões do uso do algoritmo de Metropolis para resolver, ainda que de forma aproximada, o Problema do Caixeiro Viajante (TSP).

Nota-se uma influência grande da taxa de resfriamento para com quão perto a solução apresentada será do mínimo. Com poucos pontos, essa influência é menor, de forma que até uma taxa como 0.8 é suficiente para encontrar a solução. Essa maior precisão vem ao custo de mais computações, o que normalmente implica em mais tempo.

Além disso, em problemas maiores, com mais cidades, os parâmetros de temperatura, sobretudo inicial e final, tem uma grande importância para obtermos resultados razoáveis com mais amostras mas mantendo a taxa de resfriamento menor.
