# Funções Iniciais

```python
def lcFunct(a, x_j, c, m):
    x_j1 = a*x_j + c
    x_j1 = x_j1 % m
    return x_j1

def lcGen(a, x_j0, c, m, n):
    numbers = []
    numbers.append(x_j0)
    last = x_j0
    for i in range(n):
        next = lcFunct(a, last, c, m)
        last = next
        numbers.append(last)
    return numbers

def lcGenPlot(a, x_j0, c, m, n):
    numbersLast = []
    numbersNext = []

    numbersLast.append(x_j0)
    last = x_j0
    for i in range(n):
        next = lcFunct(a, last, c, m)
        numbersNext.append(next)
        last = next
        numbersLast.append(last)
    numbersLast.pop()
    plt.scatter(numbersLast,numbersNext)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(lcGen(61807, 3141549, 0, 2**31-1,10))
```

# Item a

```python
N_list = [1000,10**4,10**5,10**6]

for N in N_list:
    numbers = lcGen(61807, 3141549, 0, 2**31-1,N)

    with plt.style.context("seaborn-whitegrid"):
        plt.hist(numbers,bins=100,edgecolor='black',linewidth=.2);
        plt.title(f"Linear Congruential - N={N}")
        plt.show()
```

```python
with plt.style.context("seaborn-whitegrid"):
    lcGenPlot(61807, 3141549, 0, 2**31-1,10**3)
```

```python
N_list = [1000,10**4,10**5,10**6]

for N in N_list:
    numbers = lcGen(61807, 3141549, 0, 2**31-1,N)
    print(f'N={N}: Mean={np.mean(numbers)}, Var={np.var(numbers)}')
```

# Built-in random

```python
from random import randint

N_list = [1000,10**4,10**5,10**6]
for N in N_list:
    numbers = []
    for i in range(N):
        numbers.append(randint(0,max(10000,N)))

    with plt.style.context("seaborn-whitegrid"):
        plt.hist(numbers,bins=100,edgecolor='black',linewidth=.2);
        plt.title(f"Linear Congruential - N={N}")
        plt.show()
```

```python
N_list = [10**3]
for N in N_list:
    numbersLast = []
    numbersNext = []

    num = randint(0,max(10000,N))
    for i in range(N):
        numbersLast.append(num)
        num = randint(0,max(10000,N))
        numbersNext.append(num)

    with plt.style.context("seaborn-whitegrid"):
        plt.scatter(numbersLast,numbersNext)
        plt.show()
```
